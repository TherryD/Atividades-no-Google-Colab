```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = {
    'Produto':['Milho','soja','Trigo','Milho','Soja','Milho'],
    'Quantidade':[10,7,5,8,6,12],
    'Preco':[5.0,8.0,6.5,5.0,8.0,5.0],
    'Regiao':['Sul','Sul','Sudeste','Sul','Nordeste','Sul']
}
df = pd.DataFrame(dados)
```

```python
df['Receita'] = df['Quantidade'] * df['Preco']
```

```python
df
```

```python
df.sort_values(by='Receita',ascending=False)
```

```python
df.groupby('Produto')['Receita'].sum()
```

```python
df.groupby('Produto')['Quantidade'].mean()
```

```python
df.loc[2,'Preco'] = None
```

```python
df.isnull().sum()
```

```python
df['Preco'] = df['Preco'].fillna(df['Preco'].mean())
```

```python
df.isnull().sum()
```

```python
receita_produto = df.groupby('Produto')['Receita'].sum()

receita_produto.plot(kind='bar')
plt.title('Receita Total por Produto')
plt.ylabel('Receita')
plt.show()
```
