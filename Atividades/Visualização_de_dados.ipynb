```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
storage_options = {'User-Agent': 'Mozilla/5.0'}

life = pd.read_csv("https://ourworldindata.org/grapher/life-expectancy.csv", storage_options=storage_options)
gdp = pd.read_csv("https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv", storage_options=storage_options)

dados = pd.merge(life, gdp, on=["Entity", "Code", "Year", ])
dados.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dados[dados['Year'] == 2020], x="GDP per capita", y="Life expectancy")
plt.title("PIB per Capita vc Expectativa de Vida")
plt.show()
```

Gráfico Scatter Plot: países com maior PIB por capital tendem a apresentar maior expectativa de vida.

```python
brasil = dados[dados['Entity'] == 'Brazil']
plt.plot(brasil['Year'], brasil['Life expectancy'])
plt.title('Evolução da Expectativa de Vida no Brasil')
plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida')
plt.show()
```

Gráfico de Linha: Observa uma tendência de crescimento contínuo da expectativa de vida ao longo das décadas.

```python
sns.histplot(dados[dados['Year'] == 2020]['Life expectancy'], kde=True)
plt.title('Distribuição da Expectativa de Vida Global em 2020')
plt.show()
```

Histograma: Mostra que a maioria dos países do mundo já se concentra na faixa de 70 a 80 anos de expectativa de vida.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=dados[dados['Year'] == 2020], x='World region according to OWID', y='Life expectancy')

plt.title('Distribuição da Expectativa de Vida por Região em 2020')
plt.xlabel('Região')
plt.ylabel('Expectativa de Vida')
plt.xticks(rotation=45)
plt.show()
```

Boxplot: Algumas regiões têm a "caixa" muito mais alta que outras e algumas regiões possuem pontos isolados.

```python
plt.figure(figsize=(8, 6))
corr = dados[['Year', 'Life expectancy', 'GDP per capita']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlação entre Variáveis')
plt.show()
```

Heatmap: Se o número estiver próximo de 1.00, significa que há uma correlação positiva forte.

Conclusão: A análise dos dados integrados permite concluir que o desenvolvimento econômico é um dos principais motores da longevidade global, apresenta uma forte correlação positiva com a expectativa de vida. Ela é mais acentuada em países em desenvolvimento.
