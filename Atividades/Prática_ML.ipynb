```python
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)

print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
X = df[['rm', 'lstat']].values
y = df['medv'].values
```

```python
import numpy as np

X = np.array(X)
y = np.array(y)
```

```python
w = np.zeros(X.shape[1])
b = 0.0
```

```python
def modelo(X, w, b):
  return X @ w + b
```

```python
def mse(y_real, y_prev):
  return np.mean((y_real - y_prev) ** 2)
```

```python
lr = 0.001
epochs = 1000
erros = []
for i in range(epochs):
  y_prev = modelo(X, w, b)
  erro = y_prev - y
  dw = (2/len(X)) * (X.T @ erro)
  db = (2/len(X)) * np.sum(erro)
  w = w - lr * dw
  b = b - lr * db
  erros.append(mse(y, y_prev))
```

```python
import matplotlib.pyplot as plt
plt.plot(erros)
plt.title("Erro ao longo do treinamento")
plt.xlabel("Iterações")
plt.ylabel("Erro (MSE)")
plt.show()
```

```python
y_prev = modelo(X, w, b)
print("Erro final:", mse(y, y_prev))
print("Pesos:", w)
print("Bias:", b)
```

```python
nova_casa = np.array([[6, 10]])
preco_previsto = modelo(nova_casa, w, b)
print("Preço previsto:", preco_previsto)
```
