```python
movie_ratings = pd.merge(movie, users, on='movie_id')
```

```python
movie_ratings.head()
```

qual chave foi usada
R: foi usada a chave e integração foi a coluna movie_id.
por que a integração é necessária
R: o dataset de avaliação possui apenas o id do filme, assim com isso podemos saber o título e o gênero é agora podemos calcular a nota média por gênero ou o filme mais assistido

```python
ranking_popul = movie_ratings.groupby('title')['rating'].count().sort_values(ascending=False)
ranking_popul.head(10)
```

```python
ranking_quantidade = movie_ratings.groupby('title')['rating'].mean().sort_values(ascending=False)
ranking_quantidade.head(10)
```

```python
movie_ratings.groupby('title')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 Filmes com mais avaliações')
plt.ylabel('Quantidade de Avaliações')
plt.show()
```

mostra a frequência de cada filme.

```python
movie_ratings['rating'].hist(bins=11, color='salmon', edgecolor='black',grid=False)
plt.title('Distribuição Geral das Notas')
plt.xlabel('Nota (0 a 10)')
plt.ylabel('Frequência')
plt.show()
```

A concentracão das notas.

```python
media_por_ano = movie_ratings.groupby('year')['rating'].mean().reset_index()

plt.scatter(media_por_ano['year'],media_por_ano['rating'], alpha=0.5, color='purple')
plt.title('Nota Média dos Filmes por Ano de Lançamento')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Nota Média')
plt.show()
```
