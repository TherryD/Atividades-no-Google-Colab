```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temperaturas = np.array([0,24,20,23,3,21,35])
print(temperaturas)
```

```python
media = np.mean(temperaturas)
print("Média das temperaturas:", media)
```

```python
plt.plot(temperaturas)
plt.title("Temperaturas Registradas")
plt.xlabel("Medição")
plt.ylabel("Temperatura")
plt.show()
```
