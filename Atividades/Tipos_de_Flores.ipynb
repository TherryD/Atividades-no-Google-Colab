```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")

iris.head()
```

```python
plt.hist(iris['petal_length'],bins=20)
plt.xlabel('Comprimento da pétala')
plt.ylabel('Frequência')
plt.title('Distribuição do comprimento da pétala')
plt.show()
```

```python
sns.kdeplot(data=iris, x='petal_length', fill=True)
plt.title('Densidade do comprimento da pétala')
plt.show()
```

```python
sns.countplot(x='species', data=iris)
plt.title('Quantidade de flores por espécie')
plt.show()
```

```python
plt.plot(iris['petal_length'])
plt.xlabel('Observação')
plt.ylabel('Comprimento da pétala')
plt.title('Sequência do comprimento da pétala')
plt.show()
```

```python
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title('Matriz de correlação')
plt.show()
```

```python
plt.scatter(iris['petal_length'],iris['petal_width'])
plt.xlabel('Comprimento da pétala')
plt.ylabel('Largura da pétala')
plt.title('Relação entre comprimento e largura da pétala')
plt.show()
```

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=iris)
plt.title('Dispersão com espécies')
plt.show()
```

```python
sns.jointplot(data=iris,x='petal_length',y='petal_width',kind='scatter')
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.stripplot(x='species',y='petal_length',data=iris)
plt.title('Distribuição do comprimento da pétala por espécie')
plt.show()
```

```python
sns.swarmplot(x='species', y='petal_length', data=iris)
plt.title('Distribuição detalhada por espécie')
plt.show()
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
plt.title('Boxplot do comprimento das pétala')
plt.show()
```

```python
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Distribuição detalhada por espécie')
plt.show()
```
