```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid_url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"

covid = pd.read_csv(covid_url)
```

```python
pop_url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"

pop =pd.read_csv(pop_url)
```

```python
countries_url = "https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv"

countries = pd.read_csv(countries_url)
```

```python
covid = [['location','date','total_cases','total_deaths']]
```

```python
covid.head()
```

```python
pop = pop[["Country Name",'Year','Value']]
```

```python
pop.columns = ['country','year','population']
```

```python
covid['date'] = pd.to_datetime(covid['date'])
covid_2021 = covid[covid['date'].dt.year == 2021]
```

```python
covid_2021.head()
```

```python
pop_2021 = pop[pop['year'] == 2021]
```

```python
covid_2021.isnull().sum()
```

```python
pop_2021.isnull().sum()
```

```python
covid_2021.duplicated().sum()
```

```python
covid = covid[covid['continent'].notna()]
```

```python
covid_2021 = covid_2021.dropna(subset=['total_cases','total_deaths'])
```

```python
covid_2021.isnull().sum()
```

```python
covid_country = covid_2021.groupby('location')[['total_cases','total_deaths']].max()
```

```python
covid_country.head()
```

```python
covid_country = covid_country.reset_index()
covid_country.columns = ['country','cases','deaths']
```

```python
dados = pd.merge(covid_country,pop_2021,on='country',how='inner')
```

```python
dados.head()
```

```python
dados['death_rate'] = dados['deaths']/dados['population']
```

```python
dados.head()
```

```python
# Reload the original 'countries' DataFrame from its URL
countries = pd.read_csv(countries_url)

# Now, select the desired columns and rename them
countries = countries[['COUNTRY','CODE']]
countries.columns = ['country','code']
```

```python
dados = pd.merge(dados,countries,on='country',how='left')
```

```python
dados.sort_values(by='cases',ascending=False).head(10)
```

```python
top10 = dados.sort_values(by='cases',ascending=False).head(10)

plt.bar(top10['country'],top10['cases'])
plt.xticks(rotation=45)
plt.title('Top 10 países com mais casos')
plt.ylabel('Casos')
plt.show()
```

```python
plt.hist(dados['deaths'],bins=30)
plt.title('Distribuição de mortes por país')
plt.show()
```

```python
plt.scatter(dados['cases'],dados['deaths'])
plt.xlabel('Total de Casos')
plt.ylabel('Total de Mortes')
plt.title('Relação entre Casos e Mortes por país')
plt.show()
```

```python
top_deaths = dados.sort_values(by='deaths',ascending=False).head(10)
plt.bar(top_deaths['country'],top_deaths['deaths'])
plt.xticks(rotation=45)
plt.title('Top 10 países com mais mortes')
plt.ylabel('Mortes')
plt.show()
```

```python
dados_flitrados = dados[dados['population'] > 100000]
```

```python
top_rate = dados_flitrados.sort_values(by='death_rate',ascending=False).head(10)
plt.bar(top_rate['country'],top_rate['death_rate'])
plt.xticks(rotation=45)
plt.title('Taxa de Mortalidade por País')
plt.ylabel('Taxa')
plt.show()
```

```python
plt.hist(dados['death_rate'],bins=40)
plt.title('Distribuição de Casos por País')
plt.xlabel('Casos')
plt.ylabel('Número de países')
plt.show()
```

```python
plt.boxplot(dados['deaths'])
plt.title('Distribuição de Mortes por País')
plt.show()
```

```python
plt.scatter(dados['population'],dados['cases'])
plt.xlabel('População')
plt.ylabel('Casos')
plt.title('Casos X População')
plt.show()
```
