```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importar as bibliotecas do python

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/winequality-red.csv"
df = pd.read_csv(url)
df.head()
```

Conjuto de dados

```python
df.info()
```

O df.info() ele confirma que o conjunto de dados está limpo.

```python
df.describe()
```

Resumo estatístico das colunas númericas.

```python
vinho = df['fixed acidity']
vinho.head()
```

Mostra uma coluno especifica dos dados.

```python
media = np.mean(vinho)
media
```

A média dos dados.

```python
plt.figure(figsize=(10,4))
plt.plot(vinho)
plt.title("Acidez do Vinho")
plt.xlabel("Dia")
plt.ylabel("vinho")
plt.show()
```

Gráfico que mostar a acidez do Vinho

```python
vinho.max()
```

A maior acidez

```python
vinho.min()
```

A menor acidez

O que os dados representam?
Estes dados representam as propriedades físico-químicas de amostras de vinho tinto.

O que você percebeu?
Com a ajuda do gráfico percebi que a acidez não é constante;ela oscila bastante entre as amostras
na média é apoximadamente 8.32 e o vinho mais ácido tem 15.9 e menos ácido tem 4.6
